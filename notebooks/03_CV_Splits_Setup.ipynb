# Standard Cross-Validation Setup (TEAM)
This notebook implements the exact `StratifiedKFold(n_splits=5, random_state=42)` split strategy.
We will generate the row indices for the 5 folds and save them to a `.pkl` file.

**Instructions for the Team:**
1. Run this notebook **ONCE** to generate `pickles/cv_fold_indices.pkl`.
2. In all your other modeling notebooks, load this pickle to get the identical Train/Validation splits.


```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
import pickle
import os

# 1. Load the clean dataset
train_path = '../pickles/train_clean.pkl'
output_path = '../pickles/cv_fold_indices.pkl'

print(f"Loading data from {train_path}...")
train = pd.read_pickle(train_path)

X = train.drop(columns=['TARGET', 'ID'], errors='ignore')
y = train['TARGET']

# 2. Define our global StratifiedKFold strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Generate the fold indices
folds = []
for fold, (train_idx, val_idx) in enumerate(cv.split(X, y)):
    folds.append({
        'fold': fold,
        'train_idx': train_idx,
        'val_idx': val_idx
    })
    print(f"Fold {fold}: Train={len(train_idx)}, Val={len(val_idx)}")

# 4. Save to a pickle file for everyone to use
os.makedirs(os.path.dirname(output_path), exist_ok=True)
with open(output_path, 'wb') as f:
    pickle.dump(folds, f)
    
print(f"\n✅ Successfully saved identical fold indices to {output_path}")
```
